```python
import numpy as np
A = np.array([[2,3],[4,4]])
B = np.array([[4,5],[6,7]])
C = A+B
print("Matrix addition:\n",C)

A = np.array([[2,3],[4,4]])
B = np.array([[4,5],[6,7]])
C = np.matmul(A,B)
print("Matrix multiplication:\n",C)

A = np.array([[2,3],[4,4]])
B = np.array([[4,5],[6,7]])
C = np.array([[3,8],[4,2]])
D = np.matmul(A,B,C)
print("Matrix multiplication:\n",D)

A = np.array([[2,3,1],[4,4,2],[2,5,8]])
B = np.array([[4,5,8],[6,7,1],[2,9,8]])
C = np.array([[3,8,2],[4,2,5],[9,1,3]])
D = np.matmul(A,B,C)
print("Matrix multiplication:\n",D)
```
